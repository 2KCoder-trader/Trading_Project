```python
import pandas as pd
import yfinance as yf
import numpy as np
import time
import matplotlib.pyplot as plt
import keyboard
from datetime import datetime
import keyboard
from IPython.display import display, clear_output
```

```python
msft = yf.Ticker("MSFT")
# grab the current time
now = datetime.now()
#find the weekday of the week 
dayofweek = now.weekday()
#find the hour of the week
hour_time = int(now.strftime("%H"))
#find the minute of the hour
minute_time = int(now.strftime("%M"))
#made for the while so it can start at 9:30
if hour_time == 9 and minute_time >= 30:
    hour_time = 10
# create figure
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
# dataframe that captures the price at that time
marketLapse = pd.DataFrame(columns=['date', 'price','index'])
marketLapse = marketLapse.set_index('date');
# dataframe that captures when the user buy sells at different prices and dates
marketAction = pd.DataFrame(columns=['date', 'price', 'status', 'color'])
marketAction = marketAction.set_index('date');
# index dates so it can be properly concatted 

# set the xlim 
firstDatetime = datetime.now();
#this is so the ylimit doesnt error
instance = pd.DataFrame([[datetime.now(), msft.info["regularMarketPrice"]]], columns=['date', 'price'])
instance = instance.set_index('date')
marketLapse = pd.concat([marketLapse, instance])
ax.plot(marketLapse.index, marketLapse['price'])

#no buy sell orders
Status = False;

try:
    while 10 <= hour_time <= 16 and 0 <= dayofweek <= 4:
        msft = yf.Ticker("MSFT")
        
        now = datetime.now()
        hour_time = int(now.strftime("%H"))
        minute_time = int(now.strftime("%M"))
        #if its 9:30
        if hour_time == 9 and minute_time >= 30:
            hour_time = 10
        #adding to the marketLapse dataframe
        instance = pd.DataFrame([[datetime.now(), msft.info["regularMarketPrice"]]], columns=['date', 'price'])
        instance = instance.set_index('date')
        marketLapse = pd.concat([marketLapse, instance])
        #for visual aid
        ax.set_xlim(firstDatetime, instance.index[0])
        ax.set_ylim(marketLapse['price'].min(), marketLapse['price'].max())
        #displaying graph
        ax.scatter(marketAction.index, marketAction['price'], color = marketAction['color'])
        ax.plot(marketLapse.index, marketLapse['price'])
        display(fig)
        clear_output(wait = True)
        plt.pause(0.5)

        t_loop = time.time() + 5
        while time.time() < t_loop:
            if keyboard.is_pressed('s') and Status:
                print("Sell")
                x = marketLapse.index[-1]
                y = marketLapse.loc[marketLapse.index[-1], "price"]
                Status = False
                instantAction = pd.DataFrame([[x, y, Status, 'blue']], columns=['date', 'price', 'status', 'color'])
                instantAction = instantAction.set_index('date')
                marketAction = pd.concat([marketAction, instantAction])
                break;

            if keyboard.is_pressed('b') and not Status:
                x = marketLapse.index[-1]
                y = marketLapse.loc[marketLapse.index[-1], "price"]
                Status = True
                instantAction = pd.DataFrame([[x, y, Status, 'red']], columns=['date', 'price', 'status', 'color'])
                instantAction = instantAction.set_index('date')
                marketAction = pd.concat([marketAction, instantAction])
                Status = True;
                break;
finally:
    print(x)
    print()
    print("The market is closed")
```
